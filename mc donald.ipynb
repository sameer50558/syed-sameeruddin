```python
# Plot bar charts for categorical columns
# Define the number of top categories to show
top_categories = 10

categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
# Count the frequency of each category
    category_counts = data[col].value_counts()
# Limit to the top categories
    top_category_counts = category_counts.head(top_categories)
# Group the rest into 'Others'
    if len(category_counts) > top_categories:
        top_category_counts['Others'] = category_counts[top_categories:].sum()
 # Plot the bar chart
    plt.figure(figsize=(10, 6))
    top_category_counts.plot(kind='bar')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Count')
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right') 
    plt.tight_layout()
    plt.show()
```

```python
# Creating a sample dataframe with categorical columns
data = pd.DataFrame({
    'Item': ['Latte with Coffee & Tea', 'Latte with F BIBLE 08/', 'McFlurry with 16. Sth', 'Biscuit', 'Biscuit'],
    'Category': ['Coffee & Tea', 'BIBLE', 'Ice Cream', 'Biscuit', 'Biscuit'],
    'Serving Size': ['36.5%', '08/', '12 fl', '7.7%', '6.2%']
})
categorical_cols = data.select_dtypes(include=['object']).columns
# Creating a figure with a specified size
fig = plt.figure(figsize=(10, 6))
# Loop through each categorical column and creating a pie chart
for i, col in enumerate(categorical_cols):
# Calculating the value counts of the column
values = data[col].value_counts()
# Creating a new subplot for each pie chart
ax = fig.add_subplot(1, len(categorical_cols), i+1)
 # Creating a pie chart of the value counts
wedges, texts, autotexts = ax.pie(values, autopct='%1.1f%%', explode=[0.1]*len(values))
# Set the title of the subplot
ax.set_title(col)
# Set the labels for each slice of the pie chart
labels = [f"{value} ({count})" for value, count in zip(values.index, values.values)]
plt.legend(wedges, labels, loc="best")
 # Set the aspect ratio of the subplot to ensure the pie chart is circular
ax.axis('equal')
plt.show()
```
